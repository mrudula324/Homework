We use supervised learning when the single outcome variable is dependent on a combination of one or more variables. the value of dependent variable can be a classification (binary or categorical) or numerical value (regression) or a combination. We use historical data first to build the models and then validate the output for accuracy. This makes SL fast for limited data.
Initially, we have to shuffle the data and split model into train,test and validation splits. We train model with training data, hyper tune the parameters using validation split and test split for testing the data.A required model is called, then we fit the model to the training data and then we predict the accuracy for the test data. Performance of the model is evaluated by calculating the differnce between predicted and the actual. For regression, root mean squared error, mean abosulte error is used as performance measurment where as for the classification model precission, accuracy is considred. 
Atlast, hyper parameter tuning is done to reduce losses and improve accuracy. Over all the model need to be fit in such a way that average loss for the entire model has to be minimized. For regression models the loss function is calculated using squared error loss, absolute error loss where as for classification we have to it is cross entropy loss and hinge loss.

Lets look into some of the supervised learning models:

**Linear Regression**: It is one of the commonly used, oldest, reliable and easy regression model used in statistics and machine learning. It finds the linear relationship between dependent and independent variables. It attempts to draw a line that closely fits the data by finding the slope and intercept. Depending on number of independent variables it can be simple linear regression or multi  linear regression. 

Downside of this model is the assumptions about the dataset. some of them includes: relationship between dependent and independent variables is linear. in real world dataset the relationships are complex, there is no relationship between independent variables, distribution of error is uniform, normal distribtion of data point when plotted between dependent and each of the independent variables. 

If the relation between independent and dependent variable fits nth degree polynomial it is considered as non-linear model. Even though it fits non-linear data from statistical view point this is considered as linear regression. Therefore, polynomial regression is considered as a special case of linear regression.

**Logistic Regression**:
It evaluates strength of relationship between dependent variable with independent variables using logistic/ sigmoid functions. Though the name suggests its regression, it is a simple classification algorithm that can be easily applied to multiple classes. Occurrence of an event (Probability) is determined using a logit function.

**KNN**:
This classification model does not make any assumptions about data. It does not require any training, so all data is used only during testing phase. so it is called simple instace-based learning algorithm. Thus is slow and uses a lot of memory as it goes through each data point. As the number of dimensions increase , over fitting initiates. Identifying number of neighbors (k) is critical for success of this classifier. less k value will lead to lower bias but higher variance. high value for k leads to low variance (smooth boundaries) but higher bias. high k values also impacts performance. since there is no easy formula to find optimal k value, Elbow method is used to get a good k value. Dimensionality reduction and filling missing values will help in improving performance and accuracy.

**SVM** : 
SVM algorithm determines best decision boundary between vectors that will help to classify data points.its advantage is, it can be applied to any kind of data. This algorithm is very good with all kinds of data (structured, semi-structured, unstructured). it works very well when underlying data is sparse and number of dimensions are high (including when number of samples are low). Chances of overfitting a model is less. 

**Decission Tree**: 
It is one of the popular supervised learning model used for classification and regression analysis. But is it considered to be more effective with classification.
It uses tree representation where each node corresponds to an attribute and leaf node to a class label. Primary challenge with this model is identifying the root node attribute, and attributes at each of the decision nodes. 
Some of the criteria that help in making attribute selection are entropy, gini entropy, variance, chi-square.
Accuracy of DT depends on entropy, a measure of data purity. the lower the entropy value the better for DT to deliniate the boundaries. If a decision cannot be made at 
node (decision node) it is further split until a decisive value is reached. 
Gini impurity is another measurement used within DT to decide whether to split the node. Entropy involves log function making it slow which calculating.
if the entropy is less amount of information availability is high and viceversa. Tree based model is selected when the dataset is non-linear and relationships between input and target variables are complex.

**Random Forest**: 
It is one of the simplest Supervised learning model that can be used both for classification and regression problems. Accuracy of this model output is high as it makes use of multiple decision trees and then merges them. Random Forest algorithm generates decision trees on randomly selected data set and searches for best features within it, then merges the prediction of each tree and makes decision based on majority. As a part of training the model, this algorithm assigns weighted average factor to each of the feature thus making the model robust.

It differs from DT, as DT goes through the training dataset it formulates set of rules to process and do prediction. RF randomly picks the observations and features to build several DT and then averages the results. If the DT is deep, the model might be an overfit. with RF, chances of Overfit are less. With fine tuning of hyper parameters RF accuracy and speed can be managed. Its performance for large dataset (with lots of DTs) has bad performance. 

**Neural network models:**
Neural networks are series of algorithm that mimic the human brain. Inputs are received as incoming function,makes a non-linear decission using activation funcation and delivers output through output function. All these are organiszed in different layers. Based on the complexity of the algrothim we can change the hidden layers. Information can flow in two ways. Feedforward networks travels only in one direction from i/p to o/p. Feedback networks in both ways forming a loop. Some of the alogithms used include gradient descent, backword propogation,feedforward and sigmoid. Though they are most widely used in classification problems this can be applied to regression models.

**Model Evaluation** :
All the classification models provide the output in terms of probability. All the results from classification models can be categorized into below 4.
a) True Positives: Actual and predicted are correct
b) True Negatives: Actual and predicted are correct
c) False Positives: Actual negative but predicted positive.
d) False Negatives: Actuals are postive predicted negative.
Based on these categories certain key metrics are calculated and applied to the model predictions depending on the requirement.
Accuracy - fraction of correct identification of results to total predictions.
Sensitivity/Recall - Fraction of positives that were correctly identified. Ex: for detecting Frauds .
Specificity -  Fraction of Negatives that are correctly identified. In diagnostic test since it is important to find number of tests as negative, it should be used when medical test result is important.
AUC uses Specificity and sensitivity which is used to measure model performance. 
Precision - Accuracy of positive predictions.
F1 score - What percent of positive predictions were correct.
macro avg - averaging the unweighted mean per class.
weighted avg - averaging the support-weighted mean per class.

Regression Models are mostly evaluated using
R-squared - propotion of variance in dependent variable that can be predicted from the independent variable.
Root Mean Squared Error - standard way of measuring error in model.
Mean Absolute Error  - average of the squares of errors. This considered as reliable measure as it penalizes the farther points with more weights and nearer points with less weight.

Unsupervised learning is employed when the requirement is not defined and identifying patterns of data is key objective. Initial data profiling is performed using anamoly detection, dimension reduction and then clustering techniques to stream line the process of model development.

Clustering is a processes of grouping data with similar features. different techniques are used to find/ calculate similarities between features of a given dataset.

**K-Means Clustering:** divides total data into groups based on k value. k is centroid of cluster. at the initial stage k values are selected randomly and the algorithm runs through the data set iteratively and assign a centroid to a group based on smallest distance (high similarity). iteration stops when there is no longer change in centroids of clusters. to determine optimal value for k, an elbow plot is made with distance on y axis and k values on x. the point where the eucledian distance is min between centroid and the points within the cluster is opted to be number of clusters.
since the algorithm has predefined number of clusters, optimal clusters for given dataset may not be accurate. 

**Hierarchial Clustering:**
It orders the rows/ columns based on similarity and makes the correlation easy. for clustering it uses the concept of proximity matrix which storesthe distances between each point. The process involves assigning all the data points to one or the other cluster based on proximity.the next step will be to merge the points with the smallest distance. this step is iterated until a single cluster is formed
Concept of Dendograms are used within heirarchial clustering to identify number of clusters.
It indicates both similarity along with their order while forming a cluster.
When the clusters are merged, a dendogram will record the distance between the merging clusters. the height of the join will be the distance between the clusters.
Threshold distance is setby drawing a horizontal line through the tallest vertical line. the number of clusters will be number of vertical lines which are being intersected by the line drawn using the threshold.

**Dimentionality Reduction**:
Dimentionality reduction is transformation of data from higher dimentional space to lower dimentional for better anlaysis. When we have too many input variables to work with we intend to use this technique. These reduction techniques tries to minimize the loss of information. It can be used for classifcaito and regression datasets. There are various linear and non-linear methods for dimentionality reduction that can be applied when data is sparse, dense. Most populariy used is PCA that can be used while working on continuous data. 

**Evaluation**: 
Evaluation of unsupervised models is not as it is for supervisised learning. clusters are evaluated based on distance between the clusters. Silhouette Coefficient: This score is higher when the clusters are well seperated and dense. Dunn index, it assumes that better clustering means that clusters are compact and well-separated from other clusters.

### 4.Perform k-means on this dataset for customer segments. Customer segments help determine what types of people buy your product, which allows you to target more people like your usual customers. Which trying to perform customer segmentation based on their purchases i have not considered all the columns. Only columns related to customerid, invoiceno and total purchase cost were taken.

```python
# Group by customerid:
ecom_df1=ecom_df.groupby('CustomerID').agg({'InvoiceNo':'count','tot_cost': 'sum'})
ecom_df1.reset_index(inplace=True)
ecom_df1
```

```python
from scipy.stats import zscore
ecom_scaled=ecom_df1.apply(zscore)
ecom_scaled
```

```python
from sklearn.cluster import KMeans
sse = {}
for k in range(1,15):
      kmeans = KMeans(n_clusters=k, random_state=1)
      kmeans.fit(ecom_df1)
      sse[k] = kmeans.inertia_


# Plotting the elbow plot
plt.figure(figsize=(12,8))
plt.title('The Elbow Method')
plt.xlabel('k'); 
plt.ylabel('Sum of squared errors')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

```python
# Let us first start with K = 5
final_model=KMeans(5)
final_model.fit(ecom_scaled)
prediction=final_model.predict(ecom_scaled)

#Append the prediction 
ecom_df1["GROUP"] = prediction
ecom_scaled["GROUP"] = prediction
print("Groups Assigned : \n")
ecom_df1
```

```python
ecom_cus_5 = ecom_df1.groupby(['GROUP'])
ecom_cus_5.mean()
```

```python
# Silhouette for cluster - 5:

from sklearn.metrics import silhouette_samples, silhouette_score

print(f'Silhouette Score(n=5): {silhouette_score(ecom_scaled, prediction)}')
```

As, we dont have predetermind formulae to caluculate the cluster size we are considering the silhouette score. After visual representation of elbow graph we narrowed our results to 3 to 5 clusters.
