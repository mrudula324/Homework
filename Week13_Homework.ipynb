### 3.Perform linear regression on the dataset to see if you can predict water temperature based on salinity. 


```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


X=np.array(bottle_df['Salnty']).reshape(-1,1)
y=np.array(bottle_df['T_degC']).reshape(-1,1)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
X.shape
```

```python
y.shape
```

```python
reg_bottle = LinearRegression()
reg_bottle.fit(X_train, y_train)
```

```python
y_pred=reg_bottle.predict(X_test)
```

```python
y_pred.shape
```

```python
accuracy_score=reg_bottle.score(X_test,y_test)
accuracy_score
```

```python
# In sampling score
reg_bottle.score(X_train, y_train)
```

#### Which means 24% of the variance from the training data has been explained by the linear model.

```python
#out of sample score
reg_bottle.score(X_test, y_test)
```

#### similarly, 23% of the variance has been explained by the test data.

### 4.Find your mean squared error and R2 values. Be sure to indicate what these tell you. You can import these from sklearn.metrics . 

```python
from sklearn.metrics import r2_score

r2_score(y_test,y_pred)
```

#### R^2 is a measure that tells the variance of dependent varables, that is explaned by all the variables (independent) in the regrssion model. R^2 value of 100% explains all the variations in the data. These values varies from 0 to 100 (expressed in percentages).
#### R^2 will always improve with the number of variables. It gives the measure of sucess.

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test,y_pred,squared=False)
```

#### Mean squared error tells the average squared difference between estimated values and actual values. MSE is always positive. If MSE is close to zero it means there no difference between actual and estimated, which implies over fitting of the data. Higher MSE leads to under fitting. Therefore, MSE should be such that it should be too low or too high.
