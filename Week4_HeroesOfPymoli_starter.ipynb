## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
a_bin=[0,9,14,19,24,29,34,39,99]           #with an assumption max player age to be 99 and age increment by 1 year
a_labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
age_bin=pd.cut(purchase_data["Age"],a_bin,labels=a_labels)
age_bin
```

```python
age_b_gr=purchase_data["SN"].groupby(age_bin).nunique()
age_b_gr
```

```python
bin_pur_count=purchase_data["SN"].groupby(age_bin).count()
bin_pur_count
```

```python
avg_purcha_p=purchase_data["Price"].groupby(age_bin).mean().sort_values(ascending=True)
avg_purcha_p
```

```python
tot_pur_value=purchase_data["Price"].groupby(age_bin).sum()
tot_pur_value
```

```python
avg_tot_pur_value=tot_pur_value/age_b_gr
avg_tot_pur_value
```

```python
pur_ana_age1={"Purchase Count":bin_pur_count,"Average Purchase Price":round(avg_purcha_p,2),"Total Purchase Value":tot_pur_value,"Avg Total Purchase per Person":round(avg_tot_pur_value,2)}
pur_ana_age_df=pd.DataFrame(pur_ana_age1)
pur_ana_age_df=pur_ana_age_df[["Purchase Count","Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"]]
pur_ana_age_df
```
