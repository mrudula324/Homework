```python
# Converting all text to lower:
wordList = [x.lower() for x in nltk.corpus.gutenberg.words('milton-paradise.txt')]
wordList
```

```python
# lexicographically sorted
wordList_sorted = set(x.lower() for x in wordList)
wordList_sorted
```

```python
#Remove stop words
wordList = [w for w in wordList  if w not in nltk.corpus.stopwords.words('english')]
wordList
```

```python
import string # Remove punctuation
wordList = [w for w in wordList if w not in string.punctuation]
wordList
```

```python
# Lemmatization compared to Stemming
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()


from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer(language = "english")


lem_ste_com = [word +' : '+ stemmer.stem(word) +' : '+ lemmatizer.lemmatize(word)  for word in wordList]
lem_ste_com
```

Lemmatization is used over stemming, while physical examining the comparision results it is observed that most meaningful words are derrived in lemmatization.

```python
lemmatizer = WordNetLemmatizer()
le = [lemmatizer.lemmatize(x)  for x in wordList]
le
```

```python
# Frequency distribution of words:
fdist = nltk.FreqDist(le)
fdist
```

```python
import itertools
import matplotlib.pyplot as plt

first_20=dict(itertools.islice(fdist.items(), 20))
print (first_20)

plt.bar(list(first_20.keys()), first_20.values(), color='g')
plt.xticks(rotation=90)
plt.show()
```

```python
# top most words by frequency:
top_20=fdist.most_common(20)

mylist = [key for key, val in top_20 for _ in range(val+1)]

plt.hist(mylist, bins=20)
plt.xticks(rotation=90)
plt.show()
```

Vander sentiment analysis (lines split by new line charector):

```python
nltk.download('vader_lexicon')
  
from nltk.sentiment.vader import SentimentIntensityAnalyzer
#nltk.download('vader_lexicon')
analyzer = SentimentIntensityAnalyzer()
```

```python
lines_list = nltk.corpus.gutenberg.raw('milton-paradise.txt').splitlines()
lines_list
```

```python
import pandas as pd
line_df=pd.DataFrame({'line': lines_list})
line_df=line_df[line_df['line'] != " "]
```

```python
neg_portion = []
neu_portion = []
pos_portion = []
compound_score = []
for sen in lines_list:
  #print (sen)
  sent_score = analyzer.polarity_scores(sen)
  #print (sent_score)
  neg_portion.append(sent_score['neg'])
  neu_portion.append(sent_score['neu'])
  pos_portion.append(sent_score['pos'])
  compound_score.append(sent_score['compound'])
```

```python
import pandas as pd
line_df=pd.DataFrame({'line': lines_list,'negative':neg_portion,'neutral':neu_portion,'positive':pos_portion,'compound':compound_score})
line_df
```

```python
# Top 5 negative lines:
line_df.sort_values(by= ['negative'],ascending=False)[:5]
```

```python
# Top 5 positive lines:
line_df.sort_values(by= ['positive'],ascending=False)[:5]
```

```python
# Top 5 neutral lines:
line_df.sort_values(by= ['neutral'],ascending=False)[:5]
```

Vander sentiment analysis ( by sentences ):

```python
from nltk import tokenize
analyzer = SentimentIntensityAnalyzer()
sen_list = tokenize.sent_tokenize(nltk.corpus.gutenberg.raw('milton-paradise.txt'))

neg_portion = []
neu_portion = []
pos_portion = []
compound_score = []
for i,sen in enumerate(sen_list):
  #print (sen[:])
  sent_score = analyzer.polarity_scores(sen[:])
  #print (sent_score)
  neg_portion.append(sent_score['neg'])
  neu_portion.append(sent_score['neu'])
  pos_portion.append(sent_score['pos'])
  compound_score.append(sent_score['compound'])


sen_df=pd.DataFrame({'line': sen_list,'negative':neg_portion,'neutral':neu_portion,'positive':pos_portion,'compound':compound_score})
sen_df
```

```python
# Top 5 neutral lines:
sen_df.sort_values(by= ['neutral'],ascending=False)[:5]
```

```python
# Top 5 positive lines:
sen_df.sort_values(by= ['positive'],ascending=False)[:5]
```

```python
# Top 5 negative lines:
sen_df.sort_values(by= ['negative'],ascending=False)[:5]
```
